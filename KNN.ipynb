```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from imutils import paths
import numpy as np
import argparse
import imutils
import cv2
import os
import pandas as pd 
import time
from skimage.io import imread
```

```python
print(os.getcwd())
# os.chdir('./cancer data')
path = os.getcwd()

df = pd.read_csv(path + '/train_labels.csv')
df
```

```python
def getOutcome(filename):
    name = ''
    for i in df['id']:
        return df.loc[df['id'] == filename.split(".")[0], 'label'].item()
```

```python
imagePath = os.getcwd() + '/train'
```

```python
def extract_color_histogram(image, bins=(8, 8, 8)):
    # extract a 3D color histogram from the HSV color space using
    # the supplied number of `bins` per channel
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    hist = cv2.calcHist([hsv], [0, 1, 2], None, bins,
        [0, 180, 0, 256, 0, 256])
    # handle normalizing the histogram if we are using OpenCV 2.4.X
    if imutils.is_cv2():
        hist = cv2.normalize(hist)
    # otherwise, perform "in place" normalization in OpenCV 3 (I
    # personally hate the way this is done
    else:
        cv2.normalize(hist, hist)
    # return the flattened histogram as the feature vector
    return hist.flatten()
```

```python
flat_data_arr = []
target_arr = []
target_var = -1
features = []
# print(os.listdir(imagePath))
ts1 = time.time()
for dirname, _, filenames in os.walk(imagePath):
    counter = 0
    for filename in filenames:
        if filename.endswith(".tif"):
            counter += 1
            path = os.path.join(dirname, filename)
            img_array=imread(path)
            flat_data_arr.append(img_array.flatten())
            print(counter, filename)
            hist = extract_color_histogram(img_array)
            target_var = getOutcome(filename)
            target_arr.append(target_var)
            features.append(hist)
ts2 = time.time()
# print(target_arr)
print("time taken : ", ts2-ts1)
```

```python
(trainRI, testRI, trainRL, testRL) = train_test_split(flat_data, target, test_size=0.25, random_state=42)
```

```python
(trainFeat, testFeat, trainLabels, testLabels) = train_test_split(features, target, test_size=0.25, random_state=42)
```

```python
print("[INFO] evaluating raw pixel accuracy...")
model = KNeighborsClassifier(n_neighbors = 1,n_jobs = -1)
model.fit(trainRI, trainRL)
acc = model.score(testRI, testRL)
print("[INFO] raw pixel accuracy: {:.2f}%".format(acc * 100))
```

```python
# train and evaluate a k-NN classifer on the histogram
# representations
print("[INFO] evaluating histogram accuracy...")
model = KNeighborsClassifier(n_neighbors=1,n_jobs=-1)
model.fit(trainFeat, trainLabels)
acc = model.score(testFeat, testLabels)
print("[INFO] histogram accuracy: {:.2f}%".format(acc * 100))
```
